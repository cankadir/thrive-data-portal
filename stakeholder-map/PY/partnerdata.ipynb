# Thrive Mapping Stakeholder Network 

## Convert Data to Graph  
If you are using the Google Sheets to update the data skip this part.

```python
import pandas as pd
path = r"C:\Git_Repo\thrive-data-portal\stakeholder-map\network-hierarchy\data.json"
df = pd.read_json( path )

df.head()
```

Natural Treasures Alliance

```python
tag = "Natural Treasures Alliance"

nta = df.dropna(subset="Tags", axis=0).copy()
nta = nta[ nta.dropna(subset="Tags", axis=0)['Tags'].apply( lambda x: True if tag in x else False ) ]

nta_orgs = nta.dropna( subset="Organization" , axis=0 )['Organization'].unique()

", ".join( nta_orgs )
```

Test Something Else

```python
tags_list = df.explode("Tags")
tags_list = tags_list[ tags_list['Organization'].notnull() ].copy()
tags_list['Tags'] = tags_list['Tags'].str.strip()
good_tags = ['Appalachian Digital Accelerator', 'Resilient Communities', 'Natural Treasures Alliance', 'Immersion Experiences', 'Freight Mobility Committee', '16 Strong', '76 towns', 'Thriving Communities', 'TN Placemaking Partnership', ]

# tags_list['Tags'].unique()

gr = tags_list[ tags_list['Tags'].isin(good_tags) ].groupby("Tags").agg({"Organization": list})


gr.explode(column="Organization").to_csv("tag-partner.csv", index=True)
```
